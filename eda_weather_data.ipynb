```python
df = df.rename({"Solar.R":"Solar"}, axis=1)
df
```

```python
df.isna().sum()
```

```python
mean = df["Solar"].mean().round()
df["Solar"] = df["Solar"].fillna(mean)
```

```python
df["Month"].unique()
```

```python
df["Month"] = df["Month"].replace({"May":'5'})
df     
```

```python
df["Date"] = pd.to_datetime(df[["Year","Month","Day"]])
```
